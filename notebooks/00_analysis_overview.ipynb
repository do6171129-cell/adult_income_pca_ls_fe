# Adult Income – PCA–LS–LGBM 構造解析事例

## 目的

本事例の目的は、モデル性能の改善を主眼とし、
PCAおよびLSによる構造解析を通じて特徴量の関係性を整理し、
得られた知見を特徴量設計（FE）へ反映する分析プロセスの一例を提示することである。

## 分析スコープ

本事例では、PCA・LS・LGBMを用いた分析パイプラインの構築および、
構造解析に基づく特徴量設計（FE）とモデル改善プロセスを対象とする。
なお、アルゴリズムの理論導出や高度なハイパーパラメータ最適化は本スコープ外とする。
fnlwgt はサンプル重み変数であり、測定された社会経済的特徴量ではないため、主成分分析の対象から除外した。

## 使用データ

- データセット：Adult（Census Income）

- 出典：UCI Machine Learning Repository

    URL：https://archive.ics.uci.edu/dataset/2/adult

- 分析対象：adult.data のみ（adult.test は使用しない）

- 問題設定：2値分類

- 目的変数：年収50Kドル超過か否か

- データ形式：表形式

- 特徴量：数値変数およびカテゴリ変数で構成

## 分析フロー

本事例では、LGBMによるベースライン評価を起点とし、
PCAおよびLSによる構造解析を通じた特徴量再設計とモデル改善を行う。
分析の全体フローを以下に示す。

- データ読み込み
- 前処理（欠損処理・型変換）
- LGBMによるベースライン学習・CV評価・予測可視化
- PCAによるデータ構造の把握
- LSによる線形関係の確認
- 構造解析に基づく特徴量設計（FE）
- LGBM再学習およびCV評価
